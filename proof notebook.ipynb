```python
import networkx as nx

G = nx.Graph()
```

```python
G.add_node(1)
```

```python
G.add_nodes_from([
    (2, {"color": "red"}),
    (3, {"color": "green"}),
])
```

```python
G.add_weighted_edges_from([(1, 2, 0.5), (1, 3, 0.75)])
```

```python
G.nodes
```

```python
import matplotlib.pyplot as plt

nx.draw(G, with_labels=True, font_weight='bold')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

def plot_maze(maze):
    # Assuming 'maze' is a 2D NumPy array where walls are represented by 1s and paths by 0s
    cmap = mcolors.ListedColormap(['white', 'black'])  # white for paths, black for walls
    plt.figure(figsize=(10, 10))
    plt.imshow(maze, cmap=cmap, interpolation='none')
    plt.axis('off')  # Turn off the axis
    plt.show()

# Example usage:
# This would be replaced with the actual 2D array representing your maze
maze = np.random.randint(2, size=(50, 50))
plot_maze(maze)
```
